```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
#Q1:計算標準常態分配，小於1的機率有多大?
# Python實現常態分配
#1.設定使用參數
mu=0
sigma=1
x=np.linspace(-10, 10, 500)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(0,1)')
plt.show()

cumsum_probs =  stats.norm.cdf(1)
cumsum_probs
```

```python
#Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
cumsum_probs =  stats.norm.cdf(-1,1)
cumsum_probs
```

```python
#Q3：計算 X~N(2,4)，小於 3 的機率有多大？
mu=2
sigma=4
x=np.linspace(-10, 10, 500)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(2,4)')
plt.show()

cumsum_probs =  stats.norm.cdf(3)
cumsum_probs
```
