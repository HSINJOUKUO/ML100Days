```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1.畫出 cos 圖檔，並儲存
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.savefig("filename.png",dpi=300,format="png")
```

```python
#2、給定散點圖顏色
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
plt.scatter(X,Y,s=50,c='g',alpha=0.3)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
