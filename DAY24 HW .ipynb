```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(5,5))
m=Basemap(projection='ortho',resolution=None,lat_0=50,lon_0=-100)
m.bluemarble(scale=0.5)
plt.show()
```

```python
plt.figure(figsize=(5,5))
m=Basemap(projection='lcc',resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107,width=12000000,height=9000000)
m.bluemarble(scale=0.5)
plt.show()
```

```python
m=Basemap(projection='mill',llcrnrlat = -90,llcrnrlon = -180,urcrnrlat = 90,urcrnrlon = 180,resolution='c')

#創建邊線
m.drawcoastlines()
#畫出國家，並使用線寬為2 的線條生成分界線。
m.drawcountries(linewidth=2)
#畫出州界
m.drawstates(color='b')
#畫出城市
m.drawcounties(color='darkred')

plt.title('Basemap Tutorial')
plt.show()
```
