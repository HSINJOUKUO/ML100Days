```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#設定圖形樣式 - whitegrid
# use sns.set

# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
# 畫圖
# sns.residplot
sns.residplot(x, y, lowess=True, color="g")
```

```python
# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();
sns.distplot(x, bins=13, kde=False, rug=True);
```

```python
sns.distplot(x, bins=13, kde=True, rug=True);
```
